## 2. Discrete Fourier Transforms of Various Signals

**The following are exercises from Chapter 7 of Mark Newman's Computational Physics book.**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

For all problems, take $N = 1,000$

2.1 Perform a DFT on a single period of a square wave of amplitude 1. Plot the wave and the magnitude spectrum, $|c_{k}|$.

```python
def square_filter(x):
    if x >= 0:
        return 1
    else:
        return 0

T = 5
N = 1000
t = np.linspace(0, T,  N)
y = [square_filter(i) for i in np.sin(2*np.pi*t/T)]
# or just 
# y = np.round((np.sin(2*np.pi*t/T) + 1) / 2)

fig, ax = plt.subplots()
ax.plot(t, y)
fig, ax = plt.subplots()
ax.plot(t[:501], np.fft.rfft(y))
```
